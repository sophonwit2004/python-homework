*‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô:*

```python
scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]
```

‡∏à‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÄ‡∏û‡∏∑‡πà‡∏≠:
- a) ‡πÅ‡∏™‡∏î‡∏á‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ (‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏° 2 ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á)

````python
scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]

# a) ‡πÅ‡∏™‡∏î‡∏á‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ (‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏° 2 ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á)
total = 0
count = 0

for i in scores:
    total = total + i
    count = count + 1

average = total / count

print("‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ = %.2f" % average)
````

### ‡∏Ç‡πâ‡∏≠ 8. ‡∏à‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏°‡∏à‡∏≤‡∏Å‡∏™‡∏π‡∏ï‡∏£ A = œÄr¬≤ ‡πÇ‡∏î‡∏¢‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ r = 7 ‡πÅ‡∏•‡∏∞‡πÉ‡∏ä‡πâ‡∏Ñ‡πà‡∏≤ œÄ ‡∏à‡∏≤‡∏Å math module (2 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

````python

# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
import math

r = 7
area = math.pi * r ** 2

print("‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏° =", area)

````

### ‡∏Ç‡πâ‡∏≠ 9. ‡∏à‡∏á‡πÅ‡∏õ‡∏•‡∏á‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏û‡∏ò‡∏≠‡∏ô (2 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

- a) $\frac{1 + 2 + 3}{4 + 5 + 6}$

- b) ‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Ç‡∏≠‡∏á 27 (Hint: $\sqrt[3]{a} = a^{1/3}$)

- c) $3^{5+\frac{3}{4}}$

````python
# a) 
a = (1+2+3)/(4+5+6)
print(a)
# b) 
b = 27 **(1/3)
print(b)
# c) 
c = 3**(5+3/4)
print(c)
````

### ‡∏Ç‡πâ‡∏≠ 10. ‡∏à‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç 2 ‡∏Ñ‡πà‡∏≤ ‡πÅ‡∏•‡πâ‡∏ß‡∏™‡∏•‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á (‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á 2 ‡∏ß‡∏¥‡∏ò‡∏µ) (2 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

````python
# Parallel Assignment
a = int(input("‡πÉ‡∏™‡πà‡∏Ñ‡πà‡∏≤ a: "))
b = int(input("‡πÉ‡∏™‡πà‡∏Ñ‡πà‡∏≤ b: "))

temp = a
a = b
b = temp

print("‡∏´‡∏•‡∏±‡∏á‡∏™‡∏•‡∏±‡∏ö‡∏Ñ‡πà‡∏≤")
print("a =", a)
print("b =", b)

print(f"‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà 2: a = {a}, b = {b}")
a = int(input("‡πÉ‡∏™‡πà‡∏Ñ‡πà‡∏≤ a: "))
b = int(input("‡πÉ‡∏™‡πà‡∏Ñ‡πà‡∏≤ b: "))

a, b = b, a

print("‡∏´‡∏•‡∏±‡∏á‡∏™‡∏•‡∏±‡∏ö‡∏Ñ‡πà‡∏≤")
print("a =", a)
print("b =", b)
````

### ‡∏Ç‡πâ‡∏≠ 14. ‡∏à‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏° FizzBuzz ‡πÇ‡∏î‡∏¢‡πÅ‡∏™‡∏î‡∏á‡πÄ‡∏•‡∏Ç 1-30 ‡∏ï‡∏≤‡∏°‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ (2 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

- ‡∏Å‡∏£‡∏ì‡∏µ‡∏≠‡∏∑‡πà‡∏ô‡πÜ ‡πÉ‡∏´‡πâ‡∏û‡∏¥‡∏°‡∏û‡πå‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏õ‡∏Å‡∏ï‡∏¥

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
for i in range(1, 31):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz", end=", ")
    elif i % 3 == 0:
        print("Fizz", end=", ")
    elif i % 5 == 0:
        print("Buzz", end=", ")
    else:
        print(i, end=", ")

````
